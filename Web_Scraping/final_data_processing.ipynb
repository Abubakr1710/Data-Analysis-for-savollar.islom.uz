# Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv('data_link/new_data.csv')
```

```python
#counting number of questions per section
data['section'].value_counts()
```

```python
#counting views per section
data.groupby('section')['views'].sum()
```

```python
#changing 'date' column to datetime format
data['date'] = pd.to_datetime(data['date'])
data.info()
```

```python
data['views'].sum()
```

```python
#getting year from date column
data['year'] = data['date'].dt.year
data
```

```python
#counting total views per year
data.groupby('year')['views'].sum()
```

```python
#top ten viewed questions in 2020
data_2020 = data[data['year'] == 2020]
data_2020 = data_2020.sort_values(by='views', ascending=False).head(10)
data_2020
```

```python
#printing 2020 year data
data_2020 = data[data['year'] == 2020]
data_2020
```

```python
#creating pie chart for number of questions per section
plt.figure(figsize=(10,10))
plt.pie(data['section'].value_counts(), labels=data['section'].value_counts().index, autopct='%1.1f%%')
plt.title('Number of questions per section')
plt.show()
```

```python
#creating bar chart for number of questions per section, but top 10 will be left others will be grouped as 'others'
plt.figure(figsize=(10,10))
plt.bar(data['section'].value_counts().index[:10], data['section'].value_counts()[:10])
plt.bar('others', data['section'].value_counts()[10:].sum())
plt.title('Number of questions per section')
plt.show()
```

```python
#changing pie chart to bar chart
plt.figure(figsize=(10,10))
plt.bar(data['section'].value_counts().head(10).index, data['section'].value_counts().head(10))
plt.title('Number of questions per section')
plt.show()
```

```python
#section name is too long. So I will make bigger font size for section name
plt.figure(figsize=(20,10))
plt.bar(data['section'].value_counts().head(10).index, data['section'].value_counts().head(10))
plt.title('Number of questions per section', fontsize=20)
plt.show()
```
